```python
import pandas as pd
import networkx as nx
import itertools
import numpy as np

df = pd.read_csv("data_small.csv",error_bad_lines=False, dtype = 'O', header=None,skiprows = 1)

df.head(15)


```

```python
#pd.options.display.max_seq_items = 2000
#pd.options.display.max_rows = 4000

```

```python
import networkx as nx

G = nx.Graph()

for x in range(50):
    for y in df.iloc[x, 10:120]:
        G.add_edge(x,y,color='r',weight=1, hastag=df.iloc[x,2], node_size = 1) 

colors = nx.get_edge_attributes(G,'color').values()
weights = nx.get_edge_attributes(G,'weight').values()
hastags = nx.get_edge_attributes(G,'hastag').values()
nodeSize = nx.get_edge_attributes(G,'node_size').values()
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(12,8), dpi= 150, facecolor='w', edgecolor='k')
#plt.subplot(211)
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos, 
        edge_color=colors,
        node_size = 50,
        width=list(weights),
        with_labels=False,
        node_color='lightgreen')
```
